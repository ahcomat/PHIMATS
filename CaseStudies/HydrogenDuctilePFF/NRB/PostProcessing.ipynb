```python
import h5py
import numpy as np
import meshio

from MeshManager import MeshManager

import matplotlib.pyplot as plt

#Set fonts
from matplotlib import rc, rcParams, cm
rc('font', **{'family': 'sans-serif', 'sans-serif': ['Arial']})
```

```python
# Plotting function
def PlotLine(y = 2*56.25/25.4, x = 1.5*56.25/25.4, dpi=100):
    fig, ax = plt.subplots(figsize=(y, x), dpi=dpi, tight_layout=True)
    return ax
```

```python
# Read nodes and node coordinate from mesh file
# Element name
elementName = "quad"  		# meshio compatible element name
mesh = MeshManager("NRB.msh", elementName)
```

```python
nSteps = 20
topY_DOFs = 2*mesh.getNodesByGroup("Top")+1

force_y = np.zeros((nSteps))
disp_y = np.zeros((nSteps))
```

```python
with h5py.File("NRB.mech.out.hdf5", "a") as fh5:
    
    for iStep in range(nSteps):
        
        force_y[iStep] = fh5["Force/Step_"+str(iStep)][()][topY_DOFs].sum()
        disp_y[iStep] = fh5["Disp/Step_"+str(iStep)][()][topY_DOFs].sum()/len(topY_DOFs)
```

```python
ax = PlotLine()

ax.plot(disp_y*1e3, force_y*1e-3, "r", lw=2)


ax.set_ylabel("Force (kN)")
ax.set_xlabel("Disp (mm)")

ax.legend(frameon=False)
# plt.savefig("FD.svg", transparent=True)
```
